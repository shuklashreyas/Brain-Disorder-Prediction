```python
import os
import shutil

target_dir = "/Users/shreyas/Desktop/BrainScanning/data"
os.makedirs(target_dir, exist_ok=True)
os.environ["KAGGLEHUB_DOWNLOAD_DIR"] = target_dir
os.chdir(target_dir)

import kagglehub


path = kagglehub.dataset_download("lukechugh/best-alzheimer-mri-dataset-99-accuracy")

if path and not path.startswith(target_dir):
    try:
        shutil.move(path, target_dir)
        print("Moved dataset from cache to:", target_dir)
        path = target_dir
    except Exception as e:
        print("Could not move dataset automatically:", e)

print("Dataset downloaded to:", path)
```
